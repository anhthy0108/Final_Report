```python
# Import các thư viện cần thiết
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
from fake_useragent import UserAgent
from selenium import webdriver
from selenium.common.exceptions import WebDriverException
```

```python
# Tạo một đối tượng UserAgent để tạo user-agent giả mạo
user_agent = UserAgent()

# Tạo một tiêu đề (header) chứa user-agent giả mạo được chọn ngẫu nhiên
header = {'User-Agent': user_agent.random}

# URL cơ sở (base_url) mà yêu cầu HTTP sẽ được gửi đến
base_url = 'https://www.sciencedirect.com/browse/journals-and-books?page={page}&contentType=JL'
```

```python
# Crawl đường dẫn (URL) và tên tạp chí
journals = []
names = []

# Đường dẫn đến tệp thực thi của ChromeDriver trên máy tính 
chromedriver_path = r'C:\Users\Asus\Downloads\Final_Report\chromedriver-win64\chromedriver-win64\chromedriver.exe'

# Thiết lập tiêu đề (header) cho trình duyệt
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument(f'user-agent={header["User-Agent"]}')
chrome_options.add_argument(f'--no-sandbox')  # Thêm tham số này để tránh lỗi trên một số máy tính

# Sử dụng ChromeOptions để tạo trình duyệt
driver = webdriver.Chrome(options=chrome_options)
driver.implicitly_wait(10)

try:
    driver.get(base_url.format(page=1))
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    data = soup.find('span', class_='pagination-pages-label u-margin-s-left-from-sm u-margin-s-right-from-sm')
    page = int(data.text.split()[-1])
    print('--> Site have:', page, 'page')

    for p in range(1, page + 1):
        driver.get(base_url.format(page=p))
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        data = soup.find_all('li', class_='publication u-padding-xs-ver js-publication')
        for i in data:
            journals.append(i.find('a', class_='anchor js-publication-title anchor-default')['href'])
            names.append(i.find('a', class_='anchor js-publication-title anchor-default').text)
        print('Page:', p)
except WebDriverException as e:
    print(f"Error: {e}")
```

```python
def get_journal_data(journal_url):
    issns = []
    print_issns = []
    linking_issns = []
    eissns = []
    imf = []

    try:
        driver.get(journal_url)
        time.sleep(10)
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # Tìm các thông tin về ISSN, Print ISSN, Linking ISSN, eISSN
        copy=soup.find('div', class_='issn-container u-margin-xs-ver u-margin-m-ver-from-md u-margin-l-right u-display-inline-block')
        value = copy.find_all('div')
        for v in value[1:]:
            text = v.text
            if 'Online ISSN' in text:
                eissns.append(text.split()[-1])
            elif 'Print ISSN' in text:
                print_issns.append(text.split()[-1])
            elif 'Linking ISSN' in text:
                linking_issns.append(text.split()[-1])
            elif 'ISSN' in text:
                issns.append(text.split()[-1])

        # Kiểm tra xem danh sách ISSN, Print ISSN, Linking ISSN, eISSN có trống hay không?
        if not eissns:
            eissns.append('')
        if not print_issns:
            print_issns.append('')
        if not linking_issns:
            linking_issns.append('')
        if not issns:
            issns.append('')

        # Tìm thông tin về Impact Factor (trường hợp 1)
        imp1 = soup.find('div', class_='sc-crrsfI gXoCQi sc-dQppl kPFoEJ js-impact-factor metric')
        if imp1:
            imf.append(imp1.text.replace('Impact Factor', ''))
        else:
            # Nếu trường hợp 1 không tồn tại, kiểm tra trường hợp 2
            imp2 = soup.find('div', class_='sc-crrsfI gXoCQi sc-dQppl kppHcg js-impact-factor metric')
            if imp2:
                imf.append(imp2.text.replace('Impact Factor', ''))
            else:
                imf.append('')

    except requests.exceptions.RequestException as e:
        print(f'Error: {e}')
        # Xử lý lỗi ở đây nếu cần
        pass

    # Trả về các danh sách ISSN, Print ISSN, Linking ISSN, eISSN và Impact Factor
    return issns, print_issns, linking_issns, eissns, imf
```

```python
# Tạo dataframe
df = pd.DataFrame(columns=['STT','Tên tạp chí','ISSN','Print ISSN','Linking ISSN','eISSN','Impact Factor'])

# Tạo một danh sách chứa số thứ tự từ 1 đến số lượng hàng trong dataframe
stt_column = list(range(1, len(df) + 1))

# Thêm cột STT vào dataframe và ép kiểu sang số nguyên
df['STT'] = stt_column
df['STT'] = df['STT'].astype(int)
```

```python
for journal, name in zip(journals, names):
    issns, print_issns, linking_issns, eissns, imf = get_journal_data('https://www.sciencedirect.com'+journal)

    # Thêm dữ liệu vào DataFrame
    data_to_append = [{'STT': len(df) + 1, 'Tên tạp chí': name, 'ISSN': ', '.join(issns), 'Print ISSN': ', '.join(print_issns), 'Linking ISSN': ', '.join(linking_issns), 'eISSN': ', '.join(eissns), 'Impact Factor': ', '.join(imf)}]
    df = pd.concat([df, pd.DataFrame(data_to_append)], ignore_index=True)

    # Xử lý lỗi
    if not issns and not eissns and not imf:
        print(f'Error for journal: {name}. Retrying in a moment...')
        time.sleep(10)  # Đợi một thời gian trước khi thử lại
        issns, print_issns, linking_issns, eissns, imf = get_journal_data('https://www.sciencedirect.com'+journal)
    
    # Đóng trình duyệt
    driver.quit()    
```
